```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../Data\Churn_Modelling - Churn_Modelling.csv')
df.head()
```

```python
df.drop(columns= ['Surname'])
```

```python
df.info()
```

```python
df=df.drop(columns= ['Surname','RowNumber','CustomerId'])
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first',sparse_output=False)
oneHot = ohe.fit_transform(df[['Geography','Gender']])

col_name = ohe.get_feature_names_out()
new_df=pd.DataFrame(oneHot, columns= col_name)
```

```python
new_df.head()
```

```python
df= pd.concat((df,new_df), axis=1)
df.head()
```

```python
df=df.drop(columns= ['Geography','Gender'])
df.head()
```
