```python
df = pd.read_csv("Data/toyota.csv")
df.head()
```

```python
df.describe()
```

```python
df = df.drop(columns= ['model'])
df.head()
```

## Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()

df['transmission'] = lb.fit_transform(df['transmission'])
df.head()
```

```python
df['transmission'].value_counts()
```

```python
df['fuelType'] = lb.fit_transform(df['fuelType'])
df.head()
```

```python
df.info()
```

## OrdinalEncoder

```python
df = pd.read_csv("Data/covid_toy.csv")
df.head()
```

```python
df['city'].unique()
```

```python
df['gender'].unique()
```

```python
df = df.drop(columns= ['age', 'fever'])
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ode_gender = OrdinalEncoder(categories= [["Male", "Female"]])
df['gender'] = ode_gender.fit_transform(df[['gender']])
df.head()
```

```python
ode = OrdinalEncoder(categories= [["Mild", "Strong"], ["Kolkata", "Bangalore", "Delhi", "Mumbai"], ["Yes", "No"]])

df[["cough", "city", "has_covid"]] = ode.fit_transform(df[["cough", "city", "has_covid"]])

df.head()
```

## MinMaxScaler

```python
df = pd.read_csv("Data/toyota.csv")
df.head()
```

```python
df = df.drop(columns = ['model'])
df.head()
```

```python
df['fuelType'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ode = OrdinalEncoder(categories= [['Manual', 'Automatic', 'Semi-Auto', 'Other'], ['Petrol', 'Other', 'Hybrid', 'Diesel']])

df[['transmission', 'fuelType']] = ode.fit_transform(df[['transmission', 'fuelType']])
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()

X = minmax_scaler.fit_transform(df)

df = pd.DataFrame(X, columns= df.columns)
df.describe()
```
