```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('../Data/Algerian_forest_fires_dataset_UPDATE.csv', header=1)
df.head()
```

```python
df.info()
```

## data cleaning

```python
df[df.isnull().any(axis=1)] 
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Region']]=df[["Region"]].astype(int)
```

## removing null values

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

## removing the 122nd row


```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

## removing spaces from the column names


```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## changes the required columns as integer data type

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

## changing the other columns to float data datatype

```python
object= [feature for feature in df.columns if df[feature].dtype=='str']
object
```

```python
for i in object:
    if i != 'Classes':
       df[i]=df[i].astype('float')
```

```python
df.info()
```

# exploratory data analysis

## drop day month and year

```python
df=df.drop(['day','month','year'],axis=1)
```

```python
df.head()
```

## encoding the class feature

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

## plot density plot for all feature

```python
plt.style.use('seaborn-v0_8')
df.hist(bins=50,figsize=(20,15))
plt.show()
```
