## OneHotEncoder

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Data/covid_toy.csv")
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
## Gender = 2 ---> 2-1 =1
# city = 4 ---> 4-1 = 3
```

```python
ohe = OneHotEncoder(drop="first", sparse_output=False)
```

```python
ohe_df = ohe.fit_transform(df[['gender', 'city', 'cough', 'has_covid']])
colums_name = ohe.get_feature_names_out()
new_df = pd.DataFrame(ohe_df, columns = colums_name)
new_df.head()
```

```python
df = pd.concat((df, new_df), axis=1)
df.head()
```

```python
df = df.drop(columns= ['gender', 'cough', 'city', 'has_covid'])
df.head()
```
